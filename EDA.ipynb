```python
#https://stackoverflow.com/questions/5967500/how-to-correctly-sort-a-string-with-a-number-inside
import re

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]
```

```python
def generate_poiunt_rank(filenames,offset=1,debug=False):
    points=[]
    ranks=[]
    for i,f in enumerate(filenames):
        df=pd.read_csv(f)
#2回目
        df=df.replace("（[0-9]+）", "", regex=True)
        df=df.replace(",", "", regex=True)
        
        if(debug):
            print(f)
            print(df.head(1))

        dpoint=pd.DataFrame()
        dpoint["名前"]=df["名前"]
        dpoint[str(i+offset)+"回目"]=df["ポイント"]
        points.append(dpoint)
    
        drank=pd.DataFrame()
        drank["名前"]=df["名前"]
        drank[str(i++offset)+"回目"]=df["順位"]
        ranks.append(drank)
    return points,ranks
```

```python
def transpose_df(df:pd.DataFrame, fillvalue=0,astype="int")-> pd.DataFrame:
        dfT=df.T
        dfT.columns=dfT.iloc[0]
        dfT=dfT.fillna(fillvalue)
        dfT=dfT.drop("名前")
        dfT=dfT.astype(astype)
        return dfT
```
