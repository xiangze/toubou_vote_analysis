## 非整数タイトル

```python
char_noninteger=pd.read_csv("data/charlinst_noninteger.csv",header=[0,1,2,3])
```

```python
char_noninteger
```

```python
id_vars = list(char_noninteger.columns[:1])
value_vars = list(char_noninteger.columns[1:])
char_noninteger_tate=char_noninteger.melt(id_vars=id_vars, value_vars=value_vars)
char_noninteger_tate
```

```python
char_noninteger_tate=char_noninteger_tate[char_noninteger_tate.columns[1:]]
char_noninteger_tate.columns=["番号","年","初登場回","作品","名前"]
```

```python
char_noninteger_tate
```

```python
char_noninteger_tate.to_csv("data/char_noninteger_list.csv")
```
