# Music idをつける

人気投票の出現順(music_points_ratio)をmusicidとする

```python
import pandas as pd
```

```python
music_point_ratio=pd.read_csv("data/music_points_ratio.csv")
```

```python
music_point_ratio = music_point_ratio.rename(columns={'Unnamed: 0': 'musicid'})
music_point_ratio
```

```python
music_title=pd.read_csv("data/music_title.csv")
```

```python
music_title
```

### タイトル英語部分除去

```python
music_title["title"]=music_title["title"].replace('\u3000.*',"",regex=True)
music_title["title"]=music_title["title"].replace('～.*',"",regex=True)
music_title["title"]=music_title["title"].replace(' - .*',"",regex=True)
music_title["title"]=music_title["title"].replace('東方',"",regex=True)
music_title
```

```python
music_title["title"].unique()
```

```python
music_title=music_title.rename(columns={"title":"作品","Unnamed: 0":"order"})
music_title
```

```python
inttitles=pd.read_csv("data/integertitles.csv")
noninttitles=pd.read_csv("data/nonintegertitles.csv")
```

```python
noninttitles
```

```python
music_title=pd.merge(music_title,noninttitles,how="outer")
#music_title["人気投票"].astype(int)
music_title=music_title.fillna(0)
music_title["年"]=music_title["年"].astype(int)
music_title["人気投票"]=music_title["人気投票"].astype(int)

#music_title=music_title.rename(columns={"人気投票":"人気投票非整数"})
music_title
```

```python
music_title_merge=music_title.merge(inttitles, left_on='作品', right_on='作品',how="outer")
music_title_merge=music_title_merge.fillna(0)
music_title_merge
```

```python
music_title_merge["年"]=music_title_merge["年_x"].astype(int)+music_title_merge["年_y"].astype(int)
music_title_merge["番号"]=music_title_merge["番号_x"].astype(float)+music_title_merge["番号_y"].astype(float)
music_title_merge["人気投票"]=music_title_merge["人気投票_x"].astype(int)+music_title_merge["人気投票_y"].astype(int)
```

```python
music_title_merge
```

```python
music_title_merge=music_title_merge.drop(columns=['番号_x', '年_x', '人気投票_x', '番号_y', '年_y', '人気投票_y'])
```

```python
music_title_merge.loc[music_title_merge["作品"]=="虹色のセプテントリオン","人気投票"]=18
```

```python
music_title_merge[music_title_merge["人気投票"]<1]
```

```python
music_title_merge.loc[music_title_merge["人気投票"]<1,"人気投票"]=1
```

```python
music_point_ratio=music_point_ratio.rename(columns={"名前":"Track Name"})
```

```python
music_point_ratio_id=pd.merge(music_point_ratio,music_title_merge)
```

```python
music_point_ratio_id=music_point_ratio_id.drop_duplicates(subset = "Track Name")
```

```python
music_point_ratio_id.to_csv("data/music_point_ratio_id.csv")
```
