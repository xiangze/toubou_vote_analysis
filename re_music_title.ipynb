# Music titleの追加

ZUN氏作曲音楽リスト †

from http://thwiki.info/?%B2%BB%B3%DA%B0%EC%CD%F7

```python
url="http://thwiki.info/?%B2%BB%B3%DA%B0%EC%CD%F7"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup
```

## titleの採集

```python
titles_o=soup.find_all("h3")
titles_o=[re.sub("\u3000.*","", t.div.text) for t in titles_o]
titles_o=[re.sub("\s.*","", t) for t in titles_o]
titles_o=[re.sub("^東方","", t) for t in titles_o]
```

```python
titles_o
```

```python
soup.find_all("p")
```

```python
#body=soup.find_all("body")[0].find_all("ul")
body=soup.find_all("body")[0]
```

```python
music_lists=body.find_all("ul",class_="list1")[1:]
```

```python
print(len(titles_o))
print(len(music_lists))
```

東方怪綺談MusicRoomにある未収録作品への対処(4つ)

```python
titles=[]
for i in titles_o:
    titles.append(i)
    if(i =="幻想郷" or i =="夢時空" or i=="封魔録" or i=="靈異伝"):
        titles.append("東方怪綺談MusicRoom")
        
print(len(titles))
print(len(music_lists))
```

```python
music_lists[0]
```

```python
print(music_lists[-1])
```

```python
musics=[[s.text for s in l.find_all("li")] for l in music_lists]
musics
```

```python
music_title=pd.DataFrame()
for ms,title in zip(musics,titles):
    for m in ms:
        try:
            name,theme=m.split(" --")
        except:
            print(m,title)
            name=m
            theme=""
        music_title=music_title.append({"name":name,"theme":theme,"title":title}, ignore_index=True)
```

```python
music_title=music_title.rename(columns={"name":"Track Name"})
music_title=music_title.rename(columns={"title":"作品"})
music_title=music_title.rename(columns={"theme":"Notes"})
music_title
```

```python
music_title[music_title["作品"]=="大空魔術"]
```

```python
def cleantitle(df:pd.DataFrame,c:str):
    df[c]=df[c].replace('\u3000.*',"",regex=True)
    df[c]=df[c].replace('～.*',"",regex=True)
    df[c]=df[c].replace('〜.*',"",regex=True)
    df[c]=df[c].replace(' - .*',"",regex=True)
    df[c]=df[c].replace(' $',"",regex=True)
    df[c]=df[c].replace('　',"",regex=True)
    df[c]=df[c].replace('\(.*\)',"",regex=True)
    df[c]=df[c].replace('（.*）',"",regex=True)
    return df

def cleantouhou(df:pd.DataFrame,c:str="title"):
    df[c]=df[c].replace('^東方',"",regex=True)    
    return df
```

```python
music_title=cleantitle(music_title,"Track Name")
music_title=cleantitle(music_title,"作品")
music_title=cleantouhou(music_title,"作品")
music_title[music_title["作品"]=="紅魔郷"]
```

```python
def rename_fill_drop(df:pd.DataFrame,cn:str):
    d=df.rename(columns={cn+"_x":cn})
    d[cn]=d[cn].fillna(d[cn+"_y"])
    d=d.drop(cn+"_y",axis=1)
    return d
```

```python
music_point_ratio_re=pd.merge(music_point_ratio_re,music_noninteger_title,how="outer",on=["Track Name"])
```

```python
music_point_ratio_re=rename_fill_drop(music_point_ratio_re,"作品")
music_point_ratio_re=rename_fill_drop(music_point_ratio_re,"Notes")
music_point_ratio_re=rename_fill_drop(music_point_ratio_re,"年")
music_point_ratio_re=rename_fill_drop(music_point_ratio_re,"人気投票")

#music_point_ratio_re[music_point_ratio_re["作品"].isna()]
```
